# First plot
Now let's plot the cosine function between -1 and 1\
To do so we need:\
x values from -1 to 1, with a step of 0.1
- create an array with the numpy function "np.arange" with a subtility, you need to stop your array at 1.1, the last value is "muted".
- calculate the cosine of x
- Lastly, we use the plot function of matplotlib.pyplot library

````python
import numpy as np
from numpy import cos
import matplotlib.pyplot as plt

x=np.arange(-1, 1.1, 0.1) # create an array containing the values (-1, -1.9, ..., 1.9, 1)
y=cos(x)
plt.plot(x, y)
````

# Matplotlib options
Let's customize our plot\
You can change the linestyle, color transparency and marker type inside the plot function

````python
import numpy as np
from numpy import cos, sin
import matplotlib.pyplot as plt

""" Processing """
x=np.arange(-pi, pi+0.1, 0.1) 
y=cos(x)
y2=sin(x)
y3=0.2*x+1

""" Plotting values """
plt.plot(x, y, color='plum', linestyle='solid', marker='.', label='cosine')
plt.plot(x, y2, color='lightsteelblue', linestyle='dashed', label='sine')
plt.plot(x, y3, color='black', linestyle='', marker='*', alpha=0.3, label='0.2x+1') # alpha sets transarency

""" Naming axes"""
plt.xlabel("x values")
plt.ylabel('y values') # in python for strings you can use either "text" or ' text'

""" Title """
plt.title('My customized plot')

""" Legend """
plt.legend(loc='upper left') # set location to your legend
````

def linear_function (slope, y_intercept, x_values):

To use this function you have to call it and replace the arguments by actual number and arrays. Two possible way of calling this function:

Why? let's define another linear function but with a default argument and commentaries
```python
def linear_function (slope, y_intercept, x_values=np.arange(0,10)):
    y_values = slope*x_values+y_intercept
    return y_values
```
Now the call would be:
```python
y=linear_function(slope=2, y-intercept=3) # first possible call

```
Last thing comment inside functions, this automatically create docstring that can display automaticaly in IDE in help window:
```python
def linear_function (slope, y_intercept, x_values=np.arange(0,10)):
    """ Compute linear function
        Args : slope (float), y_intercept (float), x_values (array), default=np.arange(0,10)
        Returns y_values (array)"""
    y_values = slope*x_values+y_intercept
    return y_values
```


````python
def linear_function (slope, y_intercept, x_values=np.arange(0,10)):
    """ Computes linear function
        Args : slope (float), y_intercept (float), x_values (array), default=np.arange(0,10)
        Returns y_values (array)"""
    y_values = slope*x_values+y_intercept
    return y_values

y1=linear_function(2, 3)
y2=linear_function(-2, 3)
y3=linear_function(0, 0)

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
````

# A word on objects and library
Other libraries contains cosine, pi, handling of arrays... Let's see what happens if we use torch instead of numpy on our basic first example

````python
import torch
from torch import cos
import matplotlib.pyplot as plt

x=torch.arange(-1, 1.1, 0.1)
y=cos(x)
plt.plot(x, y)
````

Ok but what is the difference?
Let's look at the type of x created with numpy and x created with torch

````python
import numpy as np
import torch

x_numpy=np.arange(0, 100)
x_torch= torch.arange(0,100)
print ('x_numpy:', type(x_numpy))
print ('x_torch:', type(x_torch))
````

We can see the type for numpy is numpy.ndarray meaning an array created with numpy and x_torch is a torch tensor
Basically, they both store value but they are different, how?

````python
print(x_numpy)
print(x_torch)
````

You can see that they display differently. This means that they are different objects, most of the time with diferent attributes. They are disigned to acheive different goals.
numpy arrays are used in basic scientific calculations 
torch tensor are used in machine learning
Let's try if the cosine torch function handles numpy array

````python
from torch import cos
cos(x_numpy)
````

It doesn't work, the torch library was designed to do machine learning integrates torch tensor with other attributes than numpy array (gradients for instance). The message is, python is designed to handle objects. Keep in mind that for each library, there is a different type of objects disigned for specific tasks.

# for loops and while

````python
x=np.arange(0, 10)

print('forloop')
for i in range (0, len(x)): #in range will enumerate the values between a start point and an end point
    print ('i =',i)

print('\nwhile') #\n means skip line
j=0 # to make the while loop work you need to attribute a start point to j
while j<=max(x):
    print('j =',j)
    j=j+1
````
